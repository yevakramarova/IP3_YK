```python
car_sales=pd.read_csv('/Users/EvaKramarova/Downloads/car_sales.csv')
car_sales.head()
```

```python
car_sales['Price']
```

```python
car_sales['Price'] = car_sales['Price'].replace({'\$': '', ',': ''}, regex=True).astype('int64')
```

```python
car_sales.head()
```

```python
car_sales['Total Sales']=car_sales['Price'].cumsum()
```

```python
car_sales.head()
```

```python
car_sales['Sale Date']=pd.date_range(start=pd.Timestamp.today(),periods=len(car_sales)).date
```

```python
car_sales.head()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(car_sales['Price'],car_sales['Total Sales'])
plt.title('Price vs Total Sales Line Plot')
plt.xlabel('Price')
plt.ylabel('Total Sales')
```

```python
sns.scatterplot(x=car_sales['Odometer (KM)'], y=car_sales['Price'])
plt.title('Odometer vs Price Scatterplot')
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
```

```python
sns.barplot(x=car_sales['Make'], y=car_sales['Odometer (KM)'], ci=None)
```

```python
sns.countplot(x=car_sales['Make'])
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))


axes[0, 0].plot(car_sales['Price'], car_sales['Total Sales'])
axes[0, 0].set_title('Price vs Total Sales Line Plot')


sns.scatterplot(x=car_sales['Odometer (KM)'], y=car_sales['Price'], ax=axes[0, 1])
axes[0, 1].set_title('Odometer vs Price Scatterplot')


sns.barplot(x=car_sales['Make'], y=car_sales['Odometer (KM)'], ci=None, ax=axes[1, 0])
axes[1, 0].set_title('Make vs Odometer Barplot')


sns.countplot(x=car_sales['Make'], ax=axes[1, 1])
axes[1, 1].set_title('Histogram of Car Make')
```

```python
sns.boxplot(x=car_sales['Colour'], y=car_sales['Total Sales'])
```

```python
sns.lmplot(x='Total Sales', y='Odometer (KM)', hue='Make', data=car_sales, ci=None)
```

```python
sns.pairplot(car_sales[['Price', 'Odometer (KM)', 'Total Sales', 'Doors']])
```

```python
corr_matrix=car_sales[['Total Sales', 'Price', 'Odometer (KM)']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='Spectral')
```

#### Based on the heatmap, we can see a somewhat strong negative relationship between Price and Odometer (KM). We can also see a weak positive relationsip between Price and Total Sales, and weak negative ralationship between Odometer (KM) and Total Sales.
